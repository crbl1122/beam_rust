```python


import apache_beam as beam
import float_sum  # Import the pyo3 package
from beam_aggregators import ComputeAverage

from psycopg.rows import dict_row
from beam_postgres.io import ReadAllFromPostgres

# Example usage
with beam.Pipeline() as pipeline:
    pcoll = (
        pipeline
        | "Reading example records from database" >> ReadAllFromPostgres(
        "host=host_ dbname=name_db user= .... password= ....",
        "select * from test_table",
        dict_row,
        )
        | 'Compute average' >> beam.CombineGlobally(ComputeAverage())
        | 'Print result' >> beam.Map(print)
    )
```
